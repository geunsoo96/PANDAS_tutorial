```python
import pandas as pd
```

```python

school_id_list = [{'name': 'John', 'job': "teacher", 'age': 40},
                {'name': 'Nate', 'job': "teacher", 'age': 35},
                {'name': 'Yuna', 'job': "teacher", 'age': 37},
                {'name': 'Abraham', 'job': "student", 'age': 10},
                {'name': 'Brian', 'job': "student", 'age': 12},
                {'name': 'Janny', 'job': "student", 'age': 11},
                {'name': 'Nate', 'job': "teacher", 'age': None},
                {'name': 'John', 'job': "student", 'age': None}
         ]
df = pd.DataFrame(school_id_list, columns = ['name', 'job', 'age'])
df
```

```python
# Null or NaN 확인하는법
df.info()
# age컬럼에 null값을 가진 행이 두개 있다는 것을 확인할 수 있음.
```

```python
df.isna()
# null값은 True로 출력됨
# isnull함수를 써도 됨
```

```python
# Null or NaN값 변경하기
tmp = df
tmp['age'] = tmp['age'].fillna(0)
# fillna함수를 사용하여 null값을 0으로 변경하라는 문장
tmp
```

```python

school_id_list = [{'name': 'John', 'job': "teacher", 'age': 40},
                {'name': 'Nate', 'job': "teacher", 'age': 35},
                {'name': 'Yuna', 'job': "teacher", 'age': 37},
                {'name': 'Abraham', 'job': "student", 'age': 10},
                {'name': 'Brian', 'job': "student", 'age': 12},
                {'name': 'Janny', 'job': "student", 'age': 11},
                {'name': 'Nate', 'job': "teacher", 'age': None},
                {'name': 'John', 'job': "student", 'age': None}
         ]
df = pd.DataFrame(school_id_list, columns = ['name', 'job', 'age'])
df
```

```python
df['age'].fillna(df.groupby('job')['age'].transform("median"), inplace=True)
# 해석
# null값들을 중간 평균값으로 바꾸기 위한 문장
# job 컬럼들을 groupby 후에 job컬럼 그룹으로 묶인 데이터들의 평균값을 null값에 적용
# 7번째 행은 teacher이므로 teacher값들의 평균 age값이 들어가고
# 8번째 행은 student이기 때문에 student값들의 평균 age값이 들어간다.
df
```
